```python
# Stats and Prevalence:
# To know how common each symptom is. For binary features, the mean × 100 = prevalence %.
```

```python
symptom_cols = [col for col in data.columns]

# Prevalence of each symptom
symptom_summary = data[symptom_cols].mean().sort_values(ascending=False) * 100
symptom_summary = symptom_summary.round(2).reset_index()
symptom_summary.columns = ["Symptom", "Prevalence (%)"]

print(symptom_summary)
```

```python
# Tells me: “COUGH occurs in 52% of patients, TIREDNESS in 48%, etc.”
# “High prevalence symptoms = COUGH, RUNNY_NOSE, … Low prevalence = ITCHY_ etc.”
```

```python
# Value Counts for Each Symptom:
```

```python
for col in symptom_cols:
    print("\n",data[col].value_counts(normalize=True) * 100)
```

```python
# Symptom Co-occurrence:
```

```python
from itertools import combinations
from collections import Counter

co_occurrence = Counter()

for comb in combinations(symptom_cols, 2):
    pair_count = ((data[comb[0]] == 1) & (data[comb[1]] == 1)).sum()
    co_occurrence[comb] = pair_count

# Convert to DataFrame
co_occurrence_df = pd.DataFrame(co_occurrence.items(), columns=["Symptom_Pair", "Count"])
co_occurrence_df = co_occurrence_df.sort_values("Count", ascending=False).head(15)

print(co_occurrence_df)
```

```python
# To see which symptoms often appear together.
# We’ll count co-occurrence for every pair.
```

---
**Symptom_Pair** → two symptoms that appear together in patients.

**Count** → how many patients had both of these symptoms.

- For example:
**(COUGH, TIREDNESS) = 11,175 patients had both cough and tiredness.**

```python
# Checking Coorelation:
```

```python
import seaborn as sns

corr = data[symptom_cols].corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Symptom Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Checking Association rule:
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Convert to True/False
df_bool = data[symptom_cols].astype(bool)

# Find frequent itemsets
frequent_items = apriori(df_bool, min_support=0.1, use_colnames=True)

# Generate rules
rules = association_rules(frequent_items, metric="lift", min_threshold=1.0)
rules = rules.sort_values("confidence", ascending=False).head(10)

print(rules[["antecedents", "consequents", "support", "confidence", "lift"]])
```

---
**Antecedents** = “IF” part (symptom(s) on the left side of the rule)

**Consequents** = “THEN” part (symptom(s) predicted on the right side)

**Support** = % of patients who have both antecedent and consequent

- Example: (LOSS_OF_TASTE) → (COUGH) has support 0.2373 = ~23.7% of all patients have both loss of taste & cough.

**Confidence** = probability of consequent given antecedent

- Example: For patients with LOSS_OF_TASTE, 53.4% also had COUGH.

**Lift** = how much more often the rule happens compared to random chance

- **Lift > 1** = positive association (they occur together more than random).
- **Lift ~1** = no strong association.
- **Lift < 1** = negative association (rarely occur together).
